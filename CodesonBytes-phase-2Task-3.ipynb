## 3. Top 10 directors

```python
# Assuming 'director' is another categorical column
data_directors = data[data['director']!='Not Given']
directors = data_directors['director']

# Plotting the top 10 directors
plt.figure(figsize=(10, 8))
directors.value_counts().head(10).plot(kind='bar', color='lightgreen')
plt.title('Top 10 Directors', fontsize=18)
plt.xlabel('Director', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
numerical_columns = df.select_dtypes(include=[np.number]).columns

# Visualize the distribution of numerical features using box plots
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
numerical_columns = df.select_dtypes(include=[np.number]).columns

# Visualize the distribution of numerical features
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Assuming 'rating' is another categorical column
ratings = df['rating'].value_counts()
plt.figure(figsize=(12,6))
sns.barplot(x=ratings.index,  y= ratings.values)
plt.title('Top ratings of genres')
plt.xlabel('rating')
plt.ylabel('count')
plt.show()
```

```python
# Top 10 genres:
genres = df['listed_in'].str.split(', ', expand=True).stack()

# Plotting the top 10 genres
plt.figure(figsize=(10, 8))
genres.value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Genres', fontsize=18)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Count of Ratings by Movies and Shows
new_catgs = {
    'TV-PG': 'Parental Guidance',
    'TV-MA': 'Mature Audience',
    'TV-Y7-FV': 'Teens',
    'TV-Y7': 'Teens',
    'TV-14': 'Teens',
    'R': 'Mature Audience',
    'TV-Y': 'General Audience',
    'NR': 'Mature Audience',
    'PG-13': 'Teens',
    'TV-G': 'General Audience',
    'PG': 'Teens',
    'G': 'General Audience',
    'UR': 'Mature Audience',
    'NC-17': 'Mature Audience'
}

df['rating'] = df['rating'].replace(new_catgs)

plt.figure(figsize=(10, 6))

# Plotting the count of Rating by Movies and Shows
sns.countplot(x="rating", data=df, palette="pastel", order=df['rating'].value_counts().index)

plt.title("Count of Ratings by Movies and Shows")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# Number of Movies and TV Shows Released by Year (Top 15 Years)
df_15 = df.groupby(['type', 'release_year']).size().unstack().transpose().fillna(0).sort_values(by='release_year', ascending=False)

# Plotting the top 25 years with a stacked bar plot
df_15.head(15).plot.bar(stacked=True, figsize=(17, 12))

plt.title('Number of Movies and TV Shows Released by Year (Top 15 Years)')
plt.xlabel('Year')
plt.ylabel('Number of Content')
plt.show()
```

```python
# Monthly Release Patterns Over the Years
df['date_added'] = pd.to_datetime(df['date_added'])  # Convert 'date_added' to datetime
df['month'] = df['date_added'].dt.month_name()
df['year'] = df['date_added'].dt.year

# Group by year and month and create a heatmap
df_h = df.groupby(['year', 'month']).size().unstack().fillna(0)
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(15, 10))
sns.heatmap(df_h[month_order], cmap='coolwarm', annot=True, fmt='g', cbar_kws={'label': 'Number of Releases'})
plt.title('Monthly Release Patterns Over the Years')
plt.show()
```
