```python
import pandas as pd
import requests

# Public API url storing in variable 'url'
url = "https://jsonplaceholder.typicode.com/todos"

# To get data from url
response = requests.get(url)
data = response.json()

# Extrat necessary information
to_do_list = []
for to_do in data:
    to_do_list.append(
        {"UserID":to_do['userId'],
        "ID":to_do['id'],
        "Title":to_do['title'],
        "Completed":to_do['completed']
        })
    
# Creating DataFrame
to_do_data = pd.DataFrame(to_do_list)

# Save the DataFrame to a CSV file
to_do_data.to_csv('to_do_data.csv',index = False)
print("CSV Dataset created successfully")
```
